1. **What is the difference between a function and a method in Python?**

*   Function: A function is a standalone block of code that is defined independently and can be called directly by its name. It is not tied to any particular object or class. Functions typically operate on data passed to them as arguments and return a result.

Example:


    def greet(name):
        return f"Hello, {name}!"

    message = greet("Alice")
    print(message)
*   Method: A method is a function that is defined within a class and is associated with an object of that class. Methods are called on an instance of a class and can access and manipulate the data (attributes) of that specific object. The first parameter of an instance method is conventionally named self, which refers to the instance of the class itself.

Example:

    class Dog:
        def __init__(self, name):
            self.name = name

        def bark(self):
            return f"{self.name} says Woof!"

    my_dog = Dog("Buddy")
    sound = my_dog.bark()
    print(sound)



2.  **Explain the concept of function arguments and parameters in Python**.

*   **Parameters:**

  Parameters are the variables listed inside the parentheses in a function's def statement. They act as placeholders for the values that the function expects to receive when it is called.

  Parameters define the structure of the input a function can accept. They specify the names by which the function will refer to the incoming data within its body.

  Example:
        def greet(name, age): # 'name' and 'age' are parameters
        print(f"Hello, {name}! You are {age} years old.")
*   **Arguments**:

   Arguments are the actual values or expressions passed to a function when it is called. These values are assigned to the corresponding parameters in the function's definition.
  
   Arguments provide the actual data that the function will operate on during its execution.

  Example:

      greet("Alice", 30) # "Alice" and 30 are arguments

4. **What is the purpose of the `return` statement in a Python function?**

The return statement in a Python function serves two primary purposes:

*   Exiting the Function:

    When a return statement is encountered during the execution of a function, it immediately terminates the function's execution. Control is then passed back to the point in the code where the function was called. Any code within the function after the return statement will not be executed.

*   Returning a Value:

    The return statement can be used to send a value (or multiple values, as a tuple) back to the caller of the function. This returned value can be of any Python data type, such as numbers, strings, lists, dictionaries, or even objects. The caller can then store this value in a variable, use it in an expression, or pass it as an argument to another function.

    If a function does not explicitly use a return statement, or if a return statement is used without specifying a value, the function implicitly returns None.

    Example:

         def add_numbers(a, b):
             sum_result = a + b
             return sum_result  # Returns the sum of a and b

         result = add_numbers(5, 3)
         print(result)  # Output: 8

     


5. **What are iterators in Python and how do they differ from iterables?**

*   **Iterable**

    An object that can be looped over (e.g., list, tuple, str).
    It has __iter__() method.

    Example:
          nums = [1, 2, 3]      # List is iterable
          for n in nums:        # Can use in loop
              print(n)

*   **Iterator**

    An object that remembers its state and gives next item using __next__().
    It is created using iter().
    
    Example:

        nums = [1, 2, 3]
        it = iter(nums)       # Make an iterator
        print(next(it))       # 1
        print(next(it))       # 2



6. **Explain the concept of generators in Python and how they are defined**.

*   A generator is a special type of iterator that allows you to create sequences of values on demand, rather than storing the entire sequence in memory at once. This is achieved using the yield keyword within a function, which suspends the function's execution and returns a value, allowing it to be resumed later from where it left off.

 Example:

        def count_up_to(n):
           i = 1
           while i <= n:
              yield i
              i += 1

        gen = count_up_to(3)
        print(next(gen))  # 1
        print(next(gen))  # 2




7. **What are the advantages of using generators over regular functions?**

     * Memory Efficient

         Generators yield items one by one and donâ€™t store the entire result in memory.

         Regular functions return the full list, which uses more memory.

     * Faster Startup

         Generators start producing values immediately.

         Regular functions build the entire result first.

     * Lazy Evaluation

          Values are generated only when needed using next().

          Saves CPU and memory in long loops or pipelines.
     *   Infinite Sequences Possible

      Generators can represent infinite streams without memory issues.

      Example:

             def infinite_count():
                 n = 0
                 while True:
                    yield n
                    n += 1


8. **What is a lambda function in Python and when is it typically used?**

    * A lambda function is a small, anonymous function defined using the lambda keyword.
    Unlike regular functions (def), it has no name and is usually used for simple, one-line operations.

    Example:

           add = lambda x, y: x + y
           print(add(3, 5))   # Output: 8

    * Its use:

       For short, one-line functions

       When passing a function as an argument (e.g., map(), filter(), sorted())

      Example:

            nums = [3, 1, 2]
            print(sorted(nums, key=lambda x: -x))  # [3, 2, 1]

10. **What is the difference between `map()`, `reduce()`, and `filter()` functions in Python?**

*   map(function, iterable): This function applies a given function to every item in an iterable and returns an iterator that yields the results. It is used for transformation, creating a new sequence where each element is the result of applying the function to the corresponding element in the original iterable.

   Example:

            numbers = [1, 2, 3, 4]
            squared_numbers = list(map(lambda x: x * x, numbers))
            # squared_numbers will be [1, 4, 9, 16]
      

  *   filter(function, iterable): This function constructs an iterator from elements of an iterable for which a function returns True. It is used for selection, creating a new sequence containing only the elements that satisfy a specific condition.
  
  Example:
     
             numbers = [1, 2, 3, 4, 5, 6]
             even_numbers = list(filter(lambda x: x % 2 == 0, numbers))
             # even_numbers will be [2, 4, 6]

  *   reduce(function, iterable, [initializer]): This function, found in the functools module, applies a function of two arguments cumulatively to the items of an iterable, from left to right, so as to reduce the iterable to a single value. It is used for aggregation, combining all elements into a single result. An optional initializer can be provided as the starting value.

  Example:  
   
             from functools import reduce
             numbers = [1, 2, 3, 4]
             sum_of_numbers = reduce(lambda x, y: x + y, numbers)
             # sum_of_numbers will be 10

```python
##  1. Write a Python function that takes a list of numbers as input and returns the sum of all even numbers in the list.
def input_add(l):
  s = 0
  for i in l:
    if i%2==0:
      s=s+i
  return s
input_add([1,2,3,4])
```

```python
## 3. Implement a Python function that takes a list of integers and returns a new list containing the squares of each number.
def square_list(l):
  squared = []
  for i in l:
    squared.append(i**2)
  return squared
square_list([1,2,3,4])
```

```python
## 4.  Write a Python function that checks if a given number is prime or not from 1 to 200.
def is_prime(n):
    if n < 2:
        return False
    for i in range(2, int(n**0.5) + 1):
        if n % i == 0:
            return False
    return True
for num in range(1, 201):
    if is_prime(num):
        print(f"{num} is Prime")
    else:
        print(f"{num} is Not Prime")
```

```python
## 5. Create an iterator class in Python that generates the Fibonacci sequence up to a specified number of terms
def fibo(n):
  a=0
  b=1
  for i in range(n):
    yield a
    a, b = b, a+b
it = iter(fibo(7))
print(next(it))
print(next(it))
print(next(it))
print(next(it))
print(next(it))
print(next(it))
print(next(it))
```

```python
## 6. Write a generator function in Python that yields the powers of 2 up to a given exponent.
def powers_of_two(n):
    for i in range(n + 1):
        yield 2 ** i
for num in powers_of_two(5):
    print(num)
```

```python
## 7. Implement a generator function that reads a file line by line and yields each line as a string.
def read_file_lines(filename):
    with open(filename, 'r') as file:
        for line in file:
            yield line.strip()
for line in read_file_lines("example.txt"):
    print(line)
```

```python
## 8. Use a lambda function in Python to sort a list of tuples based on the second element of each tuple.

data = [(1, 4), (3, 1), (5, 2), (2, 3)]
sorted_data = sorted(data, key=lambda x: x[1])
print(sorted_data)
```

```python
## 9. Write a Python program that uses `map()` to convert a list of temperatures from Celsius to Fahrenheit.
celsius = [0, 20, 37, 100]
fahrenheit = list(map(lambda c: (c * 9/5) + 32, celsius))
print(fahrenheit)
```

```python
## 10.Create a Python program that uses `filter()` to remove all the vowels from a given string.
def remove_vowels(s):
    vowels = 'aeiouAEIOU'
    return ''.join(filter(lambda char: char not in vowels, s))
text = "Shruti Jain"
result = remove_vowels(text)
print(result)
```
